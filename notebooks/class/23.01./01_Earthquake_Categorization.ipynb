# Earthquake_Categorization


```python
# Function to categorize earthquakes
def giveAdviceForMagnitude(magnitude: float):
    match int(magnitude):
        case 0|1:  # Micro
            return 'Everything is save.'
        case 2|3:  # Minor
            return 'Minor shakes, no damage.'
        case 4:  # Light
            return 'Light damage to buildings.'
        case 5:  # Moderate
            return 'Can cause damage of varying severity.'
        case 6:  # Strong
            return 'Can cause a lot of damage in populated areas.'
        case 7:  # Major
            return 'Can cause serious damage over larger areas.'
        case _:  # 8.0 and above
            return 'Can cause serious damage in several countries.'

def queryRichterMagnitude():
    while True:
        try:
            magnitude = float(input("Enter the magnitude of the earthquake: "))
            if magnitude < 0:
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Value must be a float larger than 0.")
    return magnitude
```

```python
magnitude = queryRichterMagnitude()
print(giveAdviceForMagnitude(magnitude))
```
